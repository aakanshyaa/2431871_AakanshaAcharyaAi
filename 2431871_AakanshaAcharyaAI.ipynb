```python
# Task - 3 - Outlier Detection:
# Tasks:
# 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
# 2. Define outliers using the 1.5 Ã— IQR rule.
# 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
# in a different color.
# 4. Discuss the characteristics of these outliers and their potential impact on regional averages.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
read = pd.read_csv("/content/filtered_data.csv")
df = pd.DataFrame(read)
def outlier(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

outlier_score = outlier(df['score'])
outlier_gdp = outlier(df['Log GDP per capita'])
# print(outlier_score,outlier_gdp)
combined = outlier_score|outlier_gdp

plt.scatter(df['Log GDP per capita'],df['score'],color = 'pink',marker='o')
plt.scatter(df.loc[combined,'Log GDP per capita'],df.loc[combined,'score'],color = 'red',marker='o')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.tight_layout()
plt.show()


```

```python
# 6. Outlier Detection:
# â€¢ Identify outlier countries in both regions based on Score and GDP per Capita.
# â€¢ Plot these outliers and discuss their implications.
middle_east_score = middle_east_datas['score']
middle_east_gdp =middle_east_datas['Log GDP per capita']
SADF_score =southasia['score']
SADF_gdp =southasia['Log GDP per capita']
def outliers(col_name):
    Q1 = col_name.quantile(0.25)
    Q3 = col_name.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (col_name < lower_bound) | (col_name > upper_bound)
middle_east_outliers_scr =outliers(middle_east_score)
middle_east_outliers_gdp =outliers(middle_east_gdp)
SADF_outliers_scr = outliers(SADF_score)
SADF_outliers_gdp = outliers(SADF_gdp)
plt.figure(figsize=(10, 6))
plt.scatter( middle_east_datas['Log GDP per capita'], middle_east_datas['score'],label='Non-Outliers', c='blue', alpha=0.5)
plt.scatter(middle_east_datas[middle_east_outliers_scr]['score'],
            middle_east_datas[middle_east_outliers_scr]['Log GDP per capita'],
            c='orange', label='Score Outliers', marker='o', edgecolors='black', s=100)
plt.scatter(middle_east_datas[middle_east_outliers_gdp]['score'],
            middle_east_datas[middle_east_outliers_gdp]['Log GDP per capita'],
            c='red', label='GDP Outliers', marker='x', s=100)
plt.title("Middle East: Outliers in Score and Log GDP per Capita")
plt.xlabel("Score")
plt.ylabel("Log GDP per Capita")
plt.legend()
plt.show()
plt.figure(figsize=(10, 6))
plt.scatter(southasia['Log GDP per capita'],southasia['score'],  label='Non-Outliers', c='green', alpha=0.5)
plt.scatter(southasia[SADF_outliers_scr]['score'],
           southasia[SADF_outliers_scr]['Log GDP per capita'],
            c='orange', label='Score Outliers', marker='o', edgecolors='black', s=100)
plt.scatter(southasia[SADF_outliers_gdp]['score'],
            southasia[SADF_outliers_gdp]['Log GDP per capita'],
            c='red', label='GDP Outliers', marker='x', s=100)
plt.title("SADF: Outliers in Score and Log GDP per Capita")
plt.xlabel("Score")
plt.ylabel("Log GDP per Capita")
plt.legend()
plt.show()
```
